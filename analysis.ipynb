```python
import pandas as pd

orders = pd.read_csv("../data/olist_orders_dataset (1).csv")
items = pd.read_csv("../data/olist_order_items_dataset (1).csv")
products = pd.read_csv("../data/olist_products_dataset (1).csv")
customers = pd.read_csv("../data/olist_customers_dataset (1).csv")
reviews = pd.read_csv("../data/olist_order_reviews_dataset (1).csv")

print("All datasets loaded successfully!")
```

```python
# Convert date column
orders["order_purchase_timestamp"] = pd.to_datetime(
    orders["order_purchase_timestamp"]
)
```

```python
df = orders.merge(
    items,
    on="order_id"
)

df = df.merge(
    products,
    on="product_id"
)

df = df.merge(
    customers,
    on="customer_id"
)

df = df.merge(
    reviews,
    on="order_id"
)

print(df.shape)
df.head()
```

```python
df["revenue"] = df["price"] + df["freight_value"]
```

```python
df[["price", "freight_value", "revenue"]].head()
```

```python
df["month"] = df["order_purchase_timestamp"].dt.month_name()

df["year_month"] = (
    df["order_purchase_timestamp"]
    .dt.to_period("M")
)
```

```python
category_sales = (
    df.groupby("product_category_name")
    ["revenue"]
    .sum()
    .sort_values(ascending=False)
)

category_sales.head(10)
```

```python
monthly_sales = (
    df.groupby("year_month")
    ["revenue"]
    .sum()
)

monthly_sales.idxmax()
```

```python
region_sales = (
    df.groupby("customer_state")
    ["revenue"]
    .sum()
    .sort_values(ascending=False)
)

region_sales.head()
```

```python
region_sales.idxmax()
```

```python
aov = (
    df.groupby("year_month")
    ["revenue"]
    .mean()
)

aov
```
